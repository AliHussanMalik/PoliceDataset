# Pakistan Crime Analysis (2000-2024)\nThis notebook reads data from the SQLite database populated by the ingestion script.

```python
import pandas as pd\nimport matplotlib.pyplot as plt\nimport seaborn as sns\nfrom sqlalchemy import create_engine, text\n\nengine = create_engine('sqlite:///../data/processed/crimes.db')\nsns.set_theme(style='whitegrid')
```

```python
df = pd.read_sql(text('SELECT * FROM crime_records'), engine)\ndf.head()
```

```python
yearly = df.groupby('year', as_index=False)['crime_count'].sum()\nplt.figure(figsize=(12,5))\nsns.lineplot(data=yearly, x='year', y='crime_count', marker='o')\nplt.title('Total Crimes by Year')\nplt.xlabel('Year')\nplt.ylabel('Crime Count')\nplt.show()
```

```python
latest_year = int(df['year'].max())\ntop_latest = df[df['year'] == latest_year].sort_values('crime_count', ascending=False).head(10)\nplt.figure(figsize=(12,6))\nsns.barplot(data=top_latest, x='crime_count', y='offence', hue='crime_category', dodge=False)\nplt.title(f'Top 10 Offences in {latest_year}')\nplt.xlabel('Crime Count')\nplt.ylabel('Offence')\nplt.show()
```
